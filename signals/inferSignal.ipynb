```python
import cv2
import numpy as np

# Load the image of the junction
image = cv2.imread('signal.jpg')

# Resize the image to 50% of its original size (adjust scaling factor as needed)
scale_percent = 50  # Percent of the original size
new_width = int(image.shape[1] * scale_percent / 100)
new_height = int(image.shape[0] * scale_percent / 100)
dim = (new_width, new_height)

# Resize the image
resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)

# Get the dimensions of the resized image
height, width, _ = resized_image.shape

# Define traffic signal positions based on the new image size
signal_positions = [
    (int(width * 0.2), int(height * 0.2)),  # Top left signal
    (int(width * 0.8), int(height * 0.2)),  # Top right signal
    (int(width * 0.2), int(height * 0.8)),  # Bottom left signal
    (int(width * 0.8), int(height * 0.8)),  # Bottom right signal
]

# Initial traffic signal states (0 for Red, 1 for Green)
# You can manually control the signal states here
signal_states = [0, 1, 1, 1]  # Modify these values to control each signal (0=Red, 1=Green)

# Define colors for red and green lights
RED = (0, 0, 255)   # Red (BGR format)
GREEN = (0, 255, 0)  # Green (BGR format)

# Draw traffic signals based on their states
for i, pos in enumerate(signal_positions):
    # Signal border (black background)
    cv2.circle(resized_image, pos, 25, (0, 0, 0), -1)
    
    # Draw the light based on signal state
    if signal_states[i] == 0:
        # Red light
        cv2.circle(resized_image, pos, 15, RED, -1)
    else:
        # Green light
        cv2.circle(resized_image, pos, 15, GREEN, -1)

# Display the resized image with traffic signals
cv2.imshow('Resized Junction with Traffic Signals', resized_image)

# Wait for any key to close the image window
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import csv

# Load the image of the junction
image = cv2.imread('signal.jpg')

# Resize the image to 50% of its original size (adjust scaling factor as needed)
scale_percent = 50  # Percent of the original size
new_width = int(image.shape[1] * scale_percent / 100)
new_height = int(image.shape[0] * scale_percent / 100)
dim = (new_width, new_height)

# Resize the image
resized_image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)

# Define colors for red and green lights
RED = (0, 0, 255)   # Red (BGR format)
GREEN = (0, 255, 0)  # Green (BGR format)

# Function to draw traffic signals
def draw_traffic_signal(image, x, y, state):
    # Signal border (black background)
    cv2.circle(image, (x, y), 25, (0, 0, 0), -1)
    
    # Draw the light based on signal state
    if state == 0:
        cv2.circle(image, (x, y), 15, RED, -1)  # Red light
    else:
        cv2.circle(image, (x, y), 15, GREEN, -1)  # Green light

# Read the CSV file containing signal coordinates and states
csv_file = 'signal_coordinates.csv'
with open(csv_file, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        # Get the coordinates and state from the CSV
        x = int(row['X'])
        y = int(row['Y'])
        state = int(row['State'])
        
        # Draw the traffic signal at the specified position
        draw_traffic_signal(resized_image, x, y, state)

# Display the resized image with traffic signals
cv2.imshow('Resized Junction with Traffic Signals', resized_image)

# Wait for any key to close the image window
cv2.waitKey(0)
cv2.destroyAllWindows()
```
